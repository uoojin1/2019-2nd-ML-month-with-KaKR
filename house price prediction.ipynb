```python
### Part 1: Exploratory Data Analysis (EDA)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline

# get data into DataFrame
data = pd.read_csv('./input/train.csv')
data.head()
```

```python
data.count()
```

```python
# check null counts
data.isnull().sum()
```

```python
# histogram
sns.distplot(data['price'], kde=False)
```

```python
sns.barplot('bedrooms', 'bathrooms', data=data)
```

```python
sns.lineplot('floors', 'bedrooms', data=data)
```

```python
sns.barplot('condition', 'price', data=data)
```

```python
sns.barplot('condition', 'yr_renovated', data=data)
```

```python
data.hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
           xlabelsize=8, ylabelsize=8, grid=False)    
plt.tight_layout(rect=(0, 0, 1.2, 1.2))   
```

```python
# correlation matrix heatmap of all variables
f, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(round(data.corr(),2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)

'''
picking out some high correlations
1) price and bathroom/sqft_living/view/grade/sqft_above/sqft_living15
2) bedrooms and bathrooms/sqft_living/sqft_above
'''
```

```python
# box plot
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))
sns.boxplot(x="bathrooms", y="price", data=data)
```

```python
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))
sns.boxplot(x="bedrooms", y="price", data=data)
```

```python
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))
sns.boxplot(x="grade", y="price", data=data)
```

```python
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))
sns.lineplot(x="sqft_living", y="price", data=data)
```
