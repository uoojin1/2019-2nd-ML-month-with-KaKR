```python
print(df_train.price.skew())
print(df_train.price.kurt())
```

skewness & Kurtosis: 왜도와 첨도를 나타낸다
왜도: 왼쪽으로 치우쳐졌을수록 값이 더 크다
첨도: 값이 3에 가까울수록 정규부높에 가깝다. 3보다 작을시 분포가 더 평탄하다. 3보다 클시 분포가 더 뾰족하다.

```python
# we will use regression model, so we want to normalize the price so that it is closer to a standard distribution
fig = plt.figure(figsize = (15,10))

fig.add_subplot(1,2,1)
res = stats.probplot(df_train.price, plot=plt)

fig.add_subplot(1,2,2)
res = stats.probplot(np.log1p(df_train.price), plot=plt)
```

logarithmic transformation leads to a more normalized distribution

```python
df_train.price = np.log1p(df_train['price'])
sns.distplot(df_train.price)
print(df_train.price.skew())
print(df_train.price.kurt())
```

```python
df_train.price.describe()
```

## visualizing the attributes
변수 시각화 (반응 변수)

```python
# top 10 features that have high correlation with the target variable
import scipy as sp

cor_abs = abs(df_train.corr()) # should use spearman coefficient matrix
cor_cols = cor_abs.nlargest(n=10, columns='price').index
cor = np.array(sp.stats.spearmanr(df_train[cor_cols].values))[0]
print(cor_cols)
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.25)
sns.heatmap(cor, fmt='.2f', annot=True, square=True, xticklabels=cor_cols.values, yticklabels=cor_cols.values)
```

interestingly, latitude has something to do with house price.
